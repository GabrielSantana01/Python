```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
dados = pd.read_csv("Dados/Dados_totais.csv")
dados_genero = pd.read_csv("Dados/data_by_genres.csv")
dados_ano = pd.read_csv("Dados/data_by_year.csv")
dados.head()
```

```python
dados = dados.drop(columns=["mode","key","explicit"],axis= True)
dados_genero = dados_genero.drop(columns=["mode","key"],axis= True)
dados_ano = dados_ano.drop(columns=["mode","key"],axis= True)
```

```python
dados_ano = dados_ano[dados_ano["year"] >= 2000]
dados_ano.reset_index()
```

```python
dados["year"].unique()
```

```python

fig = px.line(dados_ano, x="year", y="loudness", markers= True, title= "Variação do loudness pelo year")
fig.show()
```
