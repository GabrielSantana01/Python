#Criando Variaveis Aula 07

```python
dados = pd.read_csv('C:/Users/gabri/Documents/python/Data-science/Alura/Python Pandas/aluguel.csv', sep = ';')
dados
```

```python
dados["Valor Bruto"] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados
```

```python
dados['Valor m2'] = (dados['Valor'] / dados['Area']).round(2)
dados
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto']/dados['Area']).round(2)
dados
```

```python
tipos = dados.Tipo
tipos.drop_duplicates(inplace = True) 
casa = ['Casa','Casa de Condomínio','Casa de Vila']
dados['Tipos agregado'] = dados['Tipo'].apply(lambda x: 'casa' if x in casa else 'Apartamento' )
''' o apply aplica a verificação em cada linha 
da coluna. o lambda aplica a condicional de comparação pra cada linha de dados['Tipo'] se tem algum valor que estaja dentro da variavel casa, se for
verdadeira dados['Tipos agregado'] recebe o valor casa, caso seja falso dados['Tipos agregado'] recebe o valor Apartamento.'''
dados
```

```python
import pandas as pd
alunos = pd.DataFrame({'Nome': ['Ary', 'Cátia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
                        'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
                        'Idade': [15, 27, 56, 32, 42, 21, 19, 35], 
                        'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6]}, 
                        columns = ['Nome', 'Idade', 'Sexo', 'Notas'])
```

```python
alunos['Notas-Média(Notas)'] = alunos['Notas'].apply(lambda x: x - alunos['Notas'].mean())
alunos
```

```python
alunos['Faixa Etária'] = alunos['Idade'].apply(lambda x: 'Menor que 20 anos' if x < 20 
        else ('Entre 20 e 40 anos' if (x >= 20 and x <= 40) else 'Maior que 40 anos'))

alunos
```

```python
m1 = 'CCcCCccCCCccCcCccCcCcCCCcCCcccCCcCcCcCcccCCcCcccCc'
m2 = 'CCCCCccCccCcCCCCccCccccCccCccCCcCccCcCcCCcCccCccCc'
m3 = 'CccCCccCcCCCCCCCCCCcccCccCCCCCCccCCCcccCCCcCCcccCC'
m4 = 'cCCccCCccCCccCCccccCcCcCcCcCcCcCCCCccccCCCcCCcCCCC'
m5 = 'CCCcCcCcCcCCCcCCcCcCCccCcCCcccCccCCcCcCcCcCcccccCc'
```

```python
eventos = {'m1': list(m1), 
            'm2': list(m2), 
            'm3': list(m3), 
            'm4': list(m4), 
            'm5': list(m5)}
moedas = pd.DataFrame(eventos)
#moedas['m5'].value_counts()
```

```python
df = pd.DataFrame(data = ['Cara', 'Coroa'], index = ['c', 'C'], columns = ['Faces'])
df
```

```python
for item in moedas:
    df = pd.concat([df, moedas[item].value_counts()], axis = 1)
df
```

```python
#toda vez que eu queira pegar valores como o total de cada atributo conforme abaixo. o indicado é criar um novo DataFrame vazio
#e concat esse nodo DataFrame com os valores do antigo DataFrame.essa concatenação abaixo pega df01 e soma com cada coluna do moedas

df1 = pd.DataFrame()
for item in moedas:
    df1 = pd.concat([df1, moedas[item].value_counts()], axis = 1)
df1
```
